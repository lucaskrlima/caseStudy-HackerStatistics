Na aposta no Empire State Building, seu prximo movimento depende do nmero obtido aps lanar os dados. Podemos codificar isso perfeitamente com uma construo if- elif- else!

```python
# Importando biblioteca numpy as np
import numpy as np

# Setando semente
np.random.seed(123)

# Setando as etapas
etapa = 50

# Jogando o dado
dado = np.random.randint(1,7)

# Construo do fluxo de controle
if dado <= 2 :
    etapa = etapa - 1
elif dado <= 5:
    etapa = etapa + 1
else :
    etapa = etapa + np.random.randint(1,7)

# imprimindo dado e etapa
print("dado: "+str(dado), "etapa: "+str(etapa))
```

```python
#Simulando um passei aleatorio
# Inicializando uma lista passeio_aleatorio
passeio_aleatorio = [0]

# loop for com 100 simulaes
for x in range(100) :
    # Set etapa: ultimo elemento do passeio aleatrio
    etapa = passeio_aleatorio [-1]

    # jogando o dado
    dado = np.random.randint(1,7)

    # Determinando a prxima etapa
    if dado <= 2:
        etapa = etapa - 1
    elif dado <= 5:
        etapa = etapa + 1
    else:
        etapa = etapa + np.random.randint(1,7)

    # append prxima etapa em passeio aleatrio
    passeio_aleatorio.append(etapa)

# Imprimindo passeio aleatrio
print(passeio_aleatorio)
```

```python
#Ajustando o cdigo
#precisamos garantir que nunca vamos ficar abaixo de 0, para isso vamos usar a funo max()

# loop for com 100 simulaes
for x in range(100) :
    # Set etapa: ultimo elemento do passeio aleatrio
    etapa = passeio_aleatorio [-1]

    # jogando o dado
    dado = np.random.randint(1,7)

    # Determinando a prxima etapa
    if dado <= 2:
        etapa = max(0, etapa - 1)# usando funo max() para garantir que a proxima etapa no fique abaixo de zero
    elif dado <= 5:
        etapa = etapa + 1
    else:
        etapa = etapa + np.random.randint(1,7)

    # append prxima etapa em passeio aleatrio
    passeio_aleatorio.append(etapa)

# Imprimindo passeio aleatrio
print(passeio_aleatorio)
```

```python
#Visualizando graficamente o caminho aleatrio
# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot passeio_aleatorio
plt.plot(passeio_aleatorio)

# Mostrando grfico
plt.show()
```

```python
#Simulando multiplos caminhos
#Vamos simular repetidas vezes passeio_aleatorio e coletar os resultados para verificar sua distribuio.
todos_passos=[]

#Simulando o passeio aleatrio 5x
for i in range(5):

    passeio_aleatorio=[0]
    for x in range(100) :

        etapa = passeio_aleatorio [-1]

        dado = np.random.randint(1,7)

        if dado <= 2:
            etapa = max(0, etapa - 1)
        elif dado <= 5:
            etapa = etapa + 1
        else:
            etapa = etapa + np.random.randint(1,7)
        passeio_aleatorio.append(etapa)

    #append passeio_aleatorio em todos_passos
    todos_passos.append(passeio_aleatorio)

# Imprimindo todos_passos
print(todos_passos)
```

```python
# Convertendo todos_passo para um NumPy array: np_tp
np_tp = np.array(todos_passos)

# Plotando np_tp e exibindo
plt.plot(np_tp)
plt.show()
```

```python
#melhorando visualizao grfica
# Transpondo np_tp: np_tp_t
np_tp_t = np.transpose(np_tp)

# Plotando np_tp_t e exibindo
plt.plot(np_tp_t)
plt.show()
```

Na descrio do case temos 0,5% de chance de cair. Isso exige outra gerao de nmeros aleatrios. Basicamente, voc pode gerar um float aleatrio entre 0 e 1. Se este valor for menor ou igual a 0,005, voc dever redefinir a etapa para 0.

```python
#Implementando a condio 0,5% de chance de cair

# Simulando passeio aleatrio em 20 simulaes
todos_passos = []
for i in range(20) :
    passeio_aleatorio = [0]
    for x in range(100) :
        etapa = passeio_aleatorio[-1]
        dado = np.random.randint(1,7)
        if dado <= 2:
            etapa = max(0, etapa - 1)
        elif dado <= 5:
            etapa = etapa + 1
        else:
            etapa = etapa + np.random.randint(1,7)

        # Implementando chance de  0.5%
        if np.random.rand() <= 0.005 :
            etapa = 0

        passeio_aleatorio.append(etapa)
    todos_passos.append(passeio_aleatorio)

# Criando e plotando np_tp_t
np_tp_t = np.transpose(np.array(todos_passos))
plt.plot(np_tp_t)
plt.show()
```

Respondendo a pergunta principal: Quais so as chances de voc chegar ao 60 degrau do Empire State Building?

Basicamente, queremos saber sobre os pontos finais de todos os passeios aleatrios que simulamos. Esses pontos finais tm uma certa distribuio que podemos visualizar com um histograma.

```python
# Simulando passeio aleatrio em 500 simulaes
todos_passos = []
for i in range(500) :
    passeio_aleatorio = [0]
    for x in range(100) :
        etapa = passeio_aleatorio[-1]
        dado = np.random.randint(1,7)
        if dado <= 2:
            etapa = max(0, etapa - 1)
        elif dado <= 5:
            etapa = etapa + 1
        else:
            etapa = etapa + np.random.randint(1,7)

        # Implementando chance de  0.5%
        if np.random.rand() <= 0.001 :
            etapa = 0
        passeio_aleatorio.append(etapa)
    todos_passos.append(passeio_aleatorio)

# Criando e plotando np_tp_t
np_tp_t = np.transpose(np.array(todos_passos))

# Selecionando uma linha em np_tp_t: ponto_final
ponto_final = np_tp_t[-1]

# Plotando histograma de ponto_final
plt.hist(ponto_final)
plt.show()
```

Calculando as probabilidades.

O histograma anterior foi criado a partir de um array NumPy ponto_final, que contm 500 inteiros. Cada inteiro representa o ponto final de um passeio aleatrio. Para calcular a chance desse ponto final ser maior ou igual a 60, voc pode contar o nmero de nmeros inteiros ponto_final maiores ou iguais a 60 e dividir esse nmero por 500, o nmero total de simulaes.


Logo, qual  a chance estimada de voc alcanar pelo menos 60 degraus de altura se jogar este jogo do Empire State Building?


Temos duas maneiras em python de calcular esta probabilidade, a primeira utilizando mtodo count_nonzero() e a segunda atravs do mtodo mean() ambos da biblioteca numpy.

```python
#opo 1
count_int = np.count_nonzero(ponto_final >= 60)
chance =  count_int/ 500
print(chance)
```

```python
#opo 2
#Podemos observar que est opo  a mais elegante.
val = np.mean(ponto_final >= 60)
print(val)
```

Concluso:

Pelo clculo de probabilidades temos 79.4% de chance de alcanar pelo menos 60 degraus de altura.
