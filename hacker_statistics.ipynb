Na aposta no Empire State Building, seu próximo movimento depende do número obtido após lançar os dados. Podemos codificar isso perfeitamente com uma construção if- elif- else!

```python
# Importando biblioteca numpy as np
import numpy as np

# Setando semente
np.random.seed(123)

# Setando as etapas
etapa = 50

# Jogando o dado
dado = np.random.randint(1,7)

# Construção do fluxo de controle
if dado <= 2 :
    etapa = etapa - 1
elif dado <= 5:
    etapa = etapa + 1
else :
    etapa = etapa + np.random.randint(1,7)

# imprimindo dado e etapa
print("dado: "+str(dado), "etapa: "+str(etapa))
```

```python
#Simulando um passei aleatorio
# Inicializando uma lista passeio_aleatorio
passeio_aleatorio = [0]

# loop for com 100 simulações
for x in range(100) :
    # Set etapa: ultimo elemento do passeio aleatório
    etapa = passeio_aleatorio [-1]

    # jogando o dado
    dado = np.random.randint(1,7)

    # Determinando a próxima etapa
    if dado <= 2:
        etapa = etapa - 1
    elif dado <= 5:
        etapa = etapa + 1
    else:
        etapa = etapa + np.random.randint(1,7)

    # append próxima etapa em passeio aleatório
    passeio_aleatorio.append(etapa)

# Imprimindo passeio aleatório
print(passeio_aleatorio)
```

```python
#Ajustando o código
#precisamos garantir que nunca vamos ficar abaixo de 0, para isso vamos usar a função max()

# loop for com 100 simulações
for x in range(100) :
    # Set etapa: ultimo elemento do passeio aleatório
    etapa = passeio_aleatorio [-1]

    # jogando o dado
    dado = np.random.randint(1,7)

    # Determinando a próxima etapa
    if dado <= 2:
        etapa = max(0, etapa - 1)# usando função max() para garantir que a proxima etapa não fique abaixo de zero
    elif dado <= 5:
        etapa = etapa + 1
    else:
        etapa = etapa + np.random.randint(1,7)

    # append próxima etapa em passeio aleatório
    passeio_aleatorio.append(etapa)

# Imprimindo passeio aleatório
print(passeio_aleatorio)
```

```python
#Visualizando graficamente o caminho aleatório
# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot passeio_aleatorio
plt.plot(passeio_aleatorio)

# Mostrando gráfico
plt.show()
```

Distribuição do passeio aleatório.

Ao jogar o dado 100x dependendo do resultado ou subimos ou descemos alguns degraus.

Logo, Qual é a chance de chegar ao 60º degrau de altura?

```python
#Simulando multiplos caminhos
#Vamos simular repetidas vezes passeio_aleatorio e coletar os resultados para verificar sua distribuição.
todos_passos=[]

#Simulando o passeio aleatório 5x
for i in range(5):

    passeio_aleatorio=[0]
    for x in range(100) :

        etapa = passeio_aleatorio [-1]

        dado = np.random.randint(1,7)

        if dado <= 2:
            etapa = max(0, etapa - 1)
        elif dado <= 5:
            etapa = etapa + 1
        else:
            etapa = etapa + np.random.randint(1,7)
        passeio_aleatorio.append(etapa)

    #append passeio_aleatorio em todos_passos
    todos_passos.append(passeio_aleatorio)

# Imprimindo todos_passos
print(todos_passos)
```

```python
# Convertendo todos_passo para um NumPy array: np_tp
np_tp = np.array(todos_passos)

# Plotando np_tp e exibindo
plt.plot(np_tp)
plt.show()
```

```python
#melhorando visualização gráfica
# Transpondo np_tp: np_tp_t
np_tp_t = np.transpose(np_tp)

# Plotando np_tp_t e exibindo
plt.plot(np_tp_t)
plt.show()
```

Na descrição do case temos 0,5% de chance de cair. Isso exige outra geração de números aleatórios. Basicamente, você pode gerar um float aleatório entre 0 e 1. Se este valor for menor ou igual a 0,005, você deverá redefinir a etapa para 0.

```python
#Implementando a condição 0,5% de chance de cair

# Simulando passeio aleatório em 20 simulações
todos_passos = []
for i in range(20) :
    passeio_aleatorio = [0]
    for x in range(100) :
        etapa = passeio_aleatorio[-1]
        dado = np.random.randint(1,7)
        if dado <= 2:
            etapa = max(0, etapa - 1)
        elif dado <= 5:
            etapa = etapa + 1
        else:
            etapa = etapa + np.random.randint(1,7)

        # Implementando chance de  0.5%
        if np.random.rand() <= 0.005 :
            etapa = 0

        passeio_aleatorio.append(etapa)
    todos_passos.append(passeio_aleatorio)

# Criando e plotando np_tp_t
np_tp_t = np.transpose(np.array(todos_passos))
plt.plot(np_tp_t)
plt.show()
```

Respondendo a pergunta principal: Quais são as chances de você chegar ao 60º degrau do Empire State Building?

Basicamente, queremos saber sobre os pontos finais de todos os passeios aleatórios que simulamos. Esses pontos finais têm uma certa distribuição que podemos visualizar com um histograma.

```python
# Simulando passeio aleatório em 500 simulações
todos_passos = []
for i in range(500) :
    passeio_aleatorio = [0]
    for x in range(100) :
        etapa = passeio_aleatorio[-1]
        dado = np.random.randint(1,7)
        if dado <= 2:
            etapa = max(0, etapa - 1)
        elif dado <= 5:
            etapa = etapa + 1
        else:
            etapa = etapa + np.random.randint(1,7)

        # Implementando chance de  0.5%
        if np.random.rand() <= 0.001 :
            etapa = 0
        passeio_aleatorio.append(etapa)
    todos_passos.append(passeio_aleatorio)

# Criando e plotando np_tp_t
np_tp_t = np.transpose(np.array(todos_passos))

# Selecionando uma linha em np_tp_t: ponto_final
ponto_final = np_tp_t[-1]

# Plotando histograma de ponto_final
plt.hist(ponto_final)
plt.show()
```

Calculando as probabilidades.

O histograma anterior foi criado a partir de um array NumPy ponto_final, que contém 500 inteiros. Cada inteiro representa o ponto final de um passeio aleatório. Para calcular a chance desse ponto final ser maior ou igual a 60, você pode contar o número de números inteiros ponto_final maiores ou iguais a 60 e dividir esse número por 500, o número total de simulações.


Logo, qual é a chance estimada de você alcançar pelo menos 60 degraus de altura se jogar este jogo do Empire State Building?


Temos duas maneiras em python de calcular esta probabilidade, a primeira utilizando método count_nonzero() e a segunda através do método mean() ambos da biblioteca numpy.

```python
#opção 1
count_int = np.count_nonzero(ponto_final >= 60)
chance =  count_int/ 500
print(chance)
```

```python
#opção 2
#Podemos observar que está opção é a mais elegante.
val = np.mean(ponto_final >= 60)
print(val)
```

Conclusão:

Pelo cálculo de probabilidades temos 79.4% de chance de alcançar pelo menos 60 degraus de altura.
